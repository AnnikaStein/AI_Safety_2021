# Evaluation of the training

## Loss over epoch

```python
import torch

import matplotlib.pyplot as plt
import mplhep as hep
plt.style.use(hep.cms.style.ROOT)

device = torch.device("cpu")
```

```python
prev_epochs_0 = 9



NUM_DATASETS = 42


weighting_method_0 = '_as_is'

all_tr_losses_0 = []
all_val_losses_0 = []



for i in range(1, prev_epochs_0+1):
    checkpoint = torch.load(f'/home/um106329/aisafety/new_march_21/models/march2021_{i}_epochs_v6_GPU_weighted_as_is_%d_datasets.pt' % NUM_DATASETS, map_location=torch.device(device))
    loss = checkpoint['loss']
    all_tr_losses_0.append(loss)
    val_loss = checkpoint['val_loss']
    all_val_losses_0.append(val_loss)
    
    
plt.figure(1,figsize=[15,7.5])
plt.plot(all_tr_losses_0,color='royalblue')
plt.plot(all_val_losses_0,color='limegreen')
plt.title(f"Training history, {prev_epochs_0} epochs ({NUM_DATASETS} files, 102853127 jets)")
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['Training loss','Validation loss'])
plt.savefig(f'/home/um106329/aisafety/new_march_21/loss_over_epoch_no_weighting_{prev_epochs_0}_files_{prev_epochs_0}_epochs.svg', format='svg', bbox_inches='tight', facecolor='w', transparent=False)
```
