## True quark flavour distributions

```python
trtarget = torch.load('/hpcwork/um106329/new_march_21/scaled/train_targets_0.pt').long().numpy()
vatarget = torch.load('/hpcwork/um106329/new_march_21/scaled/val_targets_0.pt').long().numpy()
tetarget = torch.load('/hpcwork/um106329/new_march_21/scaled/test_targets_0.pt').long().numpy()
for d in range(1, 229):
    train_targets = torch.load('/hpcwork/um106329/new_march_21/scaled/train_targets_%d.pt' % d).long().numpy()
    val_targets = torch.load('/hpcwork/um106329/new_march_21/scaled/val_targets_%d.pt' % d).long().numpy()
    test_targets = torch.load('/hpcwork/um106329/new_march_21/scaled/test_targets_%d.pt' % d).long().numpy()
    trtarget = np.concatenate([trtarget, train_targets])
    vatarget = np.concatenate([vatarget, val_targets])
    tetarget = np.concatenate([tetarget, test_targets])
```

```python
va_rel_freq = [va0count/len(vatarget), va1count/len(vatarget), va2count/len(vatarget), va3count/len(vatarget)]
print(va_rel_freq)
va_weights = [(1 - va_rel_freq[c]) for c in range(4)]
print(va_weights)
```

```python
te0count = len(tetarget[tetarget == 0])
te1count = len(tetarget[tetarget == 1])
te2count = len(tetarget[tetarget == 2])
te3count = len(tetarget[tetarget == 3])
```

```python
print([te0count, te1count, te2count, te3count])
```

```python
'''
    Derive weights for imbalanced classes during evaluation
'''
# https://towardsdatascience.com/address-class-imbalance-easily-with-pytorch-bb540497d2a6
# https://stackoverflow.com/questions/60812032/using-weightedrandomsampler-in-pytorch
test_counts = [te0count, te1count, te2count, te3count]

unnormalized_weight_vector = [sum(test_counts)/test_counts[i] for i in range(4)]
print(unnormalized_weight_vector)
weighted_sum_of_weights = sum([unnormalized_weight_vector[i]*test_counts[i] for i in range(4)])
print(weighted_sum_of_weights)
normalized_weight_vector = [unnormalized_weight_vector[i]/weighted_sum_of_weights for i in range(4)]
print(normalized_weight_vector)

weights = [unnormalized_weight_vector[tetarget[i]] for i in range(sum(test_counts))]
```

```python
print(weights[:100])
```

```python
plt.title(f'Distribution of flavour categories\nTest sample ({len(tetarget)} jets)', fontsize=20,
              y=1.03)
plt.bar(pos, [te0count, te1count, te2count, te3count])
plt.xticks(pos, categories)
plt.ylabel('Number of jets', fontsize=20)
plt.savefig(f'/home/um106329/aisafety/new_march_21/229_te_clean.svg', bbox_inches='tight', facecolor='w', transparent=False)
```

```python
te_rel_freq = [te0count/len(tetarget), te1count/len(tetarget), te2count/len(tetarget), te3count/len(tetarget)]
print(te_rel_freq)
te_weights = [(1 - te_rel_freq[c]) for c in range(4)]
print(te_weights)
```

```python
all_len = (len(trtarget)+len(vatarget)+len(tetarget))
all_0 = tr0count + va0count + te0count
all_1 = tr1count + va1count + te1count
all_2 = tr2count + va2count + te2count
all_3 = tr3count + va3count + te3count
rel_freq = [all_0/all_len, all_1/all_len, all_2/all_len, all_3/all_len]
print(rel_freq)
weights = [(1 - rel_freq[c]) for c in range(4)]
print(weights)
```

```python
weighted_samples = [(rel_freq[s] * weights[s]) for s in range(4)]
print(weighted_samples)
```

```python
all_weighted = [(rel_freq[s]*all_len * weights[s]) for s in range(4)]
print(all_weighted)
```

```python
all_categories = [all_0, all_1, all_2, all_3]
inverse_factors = [1/rel_freq[i] for i in range(4)]
inverse_factors_normalized = [inverse_factors[i] / sum(inverse_factors) for i in range(4)]
all_cat_weighted_new = [all_categories[i] * 1/rel_freq[i] for i in range(4)]
all_cat_weighted_factors = [all_cat_weighted_new[i] / all_len for i in range(4)]
all_cat_weighted_perc = [all_cat_weighted_new[i] / sum(all_cat_weighted_new) for i in range(4)]
print(all_categories)
print(rel_freq)
print(inverse_factors)
print(inverse_factors_normalized)
print(all_cat_weighted_new)
print(all_cat_weighted_factors)
print(all_cat_weighted_perc)
```

```python
all_weighted_perc = [all_weighted[i] / sum(all_weighted) for i in range(4)]
print(all_weighted_perc)
```
