## Reading files with kl_div

```python
# example file path: input_0_Jet_eta_fgsm[0, 0.01, 0.02]_186784665__ptetaflavloss_focalloss_gamma25.0_200.npy
n_test = 186784665
kl_noise_paths = [f'inputs/kl_div/input_{i}_{definitions.input_names[i]}_noise[0, 0.01, 0.02]_{n_test}__ptetaflavloss_focalloss_gamma25.0_200.npy' for i in range(67)]
kl_fgsm_paths  = [f'inputs/kl_div/input_{i}_{definitions.input_names[i]}_fgsm[0, 0.01, 0.02]_{n_test}__ptetaflavloss_focalloss_gamma25.0_200.npy' for i in range(67)]
```

```python
kl_noise = np.concatenate([np.load(path) for path in kl_noise_paths])
```

```python
kl_fgsm  = np.concatenate([np.load(path) for path in kl_fgsm_paths])
```

```python
noise_001 = kl_noise[:,0,2]
noise_002 = kl_noise[:,1,2]

fgsm_001 = kl_fgsm[:,0,2]
fgsm_002 = kl_fgsm[:,1,2]
```

```python
kl_divs = np.vstack([noise_001, noise_002, fgsm_001, fgsm_002])
```

## Plotting

```python
jetINDEX   = definitions.jetINDEX
trackINDEX = definitions.trackINDEX
svINDEX    = definitions.svINDEX
NUM_DATASETS = 278
```

```python
fig, ax = plt.subplots(figsize=(24, 10))
#hep.cms.label(loc=0)

hm = sns.heatmap(kl_divs[:,jetINDEX], cbar=True, vmin=0, vmax=1,
                 fmt='.8f', annot_kws={'size': 16, 'rotation': 90}, annot=True, 
                 square=False, cmap=plt.cm.Reds, norm=LogNorm(vmin=0.00000001, vmax=0.5), cbar_kws=dict(pad=0.01),edgecolors='k',linewidth=0.5)
ticksX = np.arange(len(jetINDEX)) + 0.5
ticksY = np.arange(4) + 0.25
ax.set_xticks(ticksX)
ax.set_xticklabels(np.array(definitions.display_names)[jetINDEX], rotation=90, fontsize=14)
ax.xaxis.set_ticks_position('none') 
ax.set_yticks(ticksY)
ax.set_yticklabels(['Noise $\sigma=0.01$', 'Noise $\sigma=0.02$','FGSM $\epsilon=0.01$', 'FGSM $\epsilon=0.02$'], rotation=90, fontsize=14)
ax.yaxis.set_ticks_position('none') 

ax.set_title(f'Kullback-Leibler divergence, Jet variables', size=20, y=1.01)
plt.savefig(f'/home/um106329/aisafety/june_21/evaluate/inputs/kl_div/noise0.01,0.02_fgsm0.01,0.02_ptetaflavloss_focalloss_gamma25.0_ep200_test{n_test}Jet.svg', bbox_inches='tight')
plt.savefig(f'/home/um106329/aisafety/june_21/evaluate/inputs/kl_div/noise0.01,0.02_fgsm0.01,0.02_ptetaflavloss_focalloss_gamma25.0_ep200_test{n_test}Jet.pdf', bbox_inches='tight')
plt.show(block=False)
time.sleep(5)
plt.close('all')
gc.collect(2)


fig, ax = plt.subplots(figsize=(24, 12))
#hep.cms.label(loc=0)

hm = sns.heatmap(kl_divs[:,trackINDEX], cbar=True, vmin=0, vmax=1,
                 fmt='.8f', annot_kws={'size': 16, 'rotation': 90}, annot=True, 
                 square=False, cmap=plt.cm.Reds, norm=LogNorm(vmin=0.00000001, vmax=0.5), cbar_kws=dict(pad=0.01),edgecolors='k',linewidth=0.5)

ticksX = np.arange(len(trackINDEX)) + 0.5
ticksY = np.arange(4) + 0.25
ax.set_xticks(ticksX)
ax.set_xticklabels(np.array(definitions.display_names)[trackINDEX], rotation=90, fontsize=14)
ax.xaxis.set_ticks_position('none') 
ax.set_yticks(ticksY)
ax.set_yticklabels(['Noise $\sigma=0.01$', 'Noise $\sigma=0.02$','FGSM $\epsilon=0.01$', 'FGSM $\epsilon=0.02$'], rotation=90, fontsize=14)
ax.yaxis.set_ticks_position('none') 

ax.set_title(f'Kullback-Leibler divergence, Track variables', size=20, y=1.02)
plt.savefig(f'/home/um106329/aisafety/june_21/evaluate/inputs/kl_div/noise0.01,0.02_fgsm0.01,0.02_ptetaflavloss_focalloss_gamma25.0_ep200_test{n_test}Track.svg', bbox_inches='tight')
plt.savefig(f'/home/um106329/aisafety/june_21/evaluate/inputs/kl_div/noise0.01,0.02_fgsm0.01,0.02_ptetaflavloss_focalloss_gamma25.0_ep200_test{n_test}Track.pdf', bbox_inches='tight')
plt.show(block=False)
time.sleep(5)
plt.close('all')
gc.collect(2)


fig, ax = plt.subplots(figsize=(24, 12))
#hep.cms.label(loc=0)

hm = sns.heatmap(kl_divs[:,svINDEX], cbar=True, vmin=0, vmax=1,
                 fmt='.8f', annot_kws={'size': 16, 'rotation': 90}, annot=True, 
                 square=False, cmap=plt.cm.Reds, norm=LogNorm(vmin=0.00000001, vmax=0.5), cbar_kws=dict(pad=0.01),edgecolors='k',linewidth=0.5)

ticksX = np.arange(len(svINDEX)) + 0.5
ticksY = np.arange(4) + 0.25
ax.set_xticks(ticksX)
ax.set_xticklabels(np.array(definitions.display_names)[svINDEX], rotation=90, fontsize=14)
ax.xaxis.set_ticks_position('none') 
ax.set_yticks(ticksY)
ax.set_yticklabels(['Noise $\sigma=0.01$', 'Noise $\sigma=0.02$','FGSM $\epsilon=0.01$', 'FGSM $\epsilon=0.02$'], rotation=90, fontsize=14)
ax.yaxis.set_ticks_position('none') 

ax.set_title(f'Kullback-Leibler divergence, Secondary Vertex variables', size=20, y=1.01)
plt.savefig(f'/home/um106329/aisafety/june_21/evaluate/inputs/kl_div/noise0.01,0.02_fgsm0.01,0.02_ptetaflavloss_focalloss_gamma25.0_ep200_test{n_test}SV.svg', bbox_inches='tight')
plt.savefig(f'/home/um106329/aisafety/june_21/evaluate/inputs/kl_div/noise0.01,0.02_fgsm0.01,0.02_ptetaflavloss_focalloss_gamma25.0_ep200_test{n_test}SV.pdf', bbox_inches='tight')
plt.show(block=False)
time.sleep(5)
plt.close('all')
gc.collect(2)
```
